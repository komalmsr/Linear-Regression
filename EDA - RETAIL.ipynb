### Checking for Missing Values

```python
data['Country'].value_counts()
```

Clearly the data is for US country only, so we can drop the 'Country' column as we dont need any analysis to be done based on it.

```python
#dropping Country column
data=data.drop('Country',axis=1)
```

```python
data
```

```python
#Checking correlation between columns visually
f,ax = plt.subplots(figsize=(9,5))
corrMatrix = data.corr()
sns.heatmap(corrMatrix, annot = True)
plt.show()
```

```python
# Summary Statistics

data.describe()
```

```python
#State wise profit
plt.figure(figsize=[24,15])
ax = sns.barplot(x="State", y="Profit", data=data, palette="Set1",)
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("States VS Profit",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("Profit",fontsize=20)
plt.tight_layout()
```

```python
#The segment which order frequently
data['Segment'].nunique()
```

```python
data_topsegment=data['Segment'].value_counts()
data_topsegment
```

```python
fig=plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
s=sns.countplot('Segment', data = data)
for s in ax.patches:
    ax.annotate('{:.0f}'.format(s.get_height()), (s.get_x()+0.15, s.get_height()+1))
plt.show()
```

```python
# Lets look at the product categories available to shop for customers
data['Category'].unique()
```

```python
#number of products in each category 
data['Category'].value_counts()
```

```python
#number of Sub-categories products are divided.
data['Sub-Category'].nunique()
```

```python
#number of products in each sub-category
data['Sub-Category'].value_counts()
```

```python
#Count of Sub-Category region wise
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.show()
```

```python
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
# plt.legend('Profit')
# plt.legend('Sales')
plt.show()
```

```python
#Lets see how sub-categories are distributed wrt to category
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data,color='y')
plt.show()
```

```python
plt.figure(figsize=(12,10))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```
